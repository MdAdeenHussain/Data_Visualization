```python
import PyPDF2

def read_pdf(file_path):
    text_data = []
    
    with open(file_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        
        for page_number, page in enumerate(reader.pages):
            text = page.extract_text()
            if text:
                text_data.append({
                    "page": page_number + 1,
                    "content": text
                })
    
    return text_data


def search_keyword(pdf_data, keyword):
    results = []
    
    for page in pdf_data:
        lines = page["content"].split("\n")
        
        for line in lines:
            if keyword.lower() in line.lower():
                results.append({
                    "page": page["page"],
                    "line": line.strip()
                })
    
    return results


# ===========================
# MAIN PROGRAM
# ===========================

file_path = "SIR_LIST.pdf"  # Replace with your PDF filename
pdf_data = read_pdf(file_path)

print("PDF Loaded Successfully!\n")

search_input = input("Enter name or keyword to search: ")

results = search_keyword(pdf_data, search_input)

if results:
    print("\nSearch Results:\n")
    for r in results:
        print(f"Page {r['page']} → {r['line']}")
else:
    print("No matching records found.")
```
